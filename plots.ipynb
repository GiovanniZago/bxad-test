```python
import numpy as np
import awkward as ak
import uproot

import mplhep as hep
import matplotlib.pyplot as plt

hep.style.use("CMS")
plt.rcParams.update({'font.size': 16})
```

```python
# feature_list = ["L1BMTFStub_hwEta", "L1BMTFStub_hwPhi", "L1BMTFStub_hwPhiB", "bunchCrossing"]
feature_list = ["nL1BMTFStub"]
```

```python
seq3_data = uproot.open("data/output_1000_seq3.root:L1BMTFStubSequences").arrays()
seq4_data = uproot.open("data/output_1000_seq4.root:L1BMTFStubSequences").arrays()
seq5_data = uproot.open("data/output_1000_seq5.root:L1BMTFStubSequences").arrays()
data = uproot.open("data/output_1000.root:Events").arrays(feature_list)
```

```python
run = True

if run:
    plt.figure(figsize=(7,5))
    hep.cms.label(label="Private Work", data=True, rlabel="Level-1 Trigger Scouring 2024 (13.6 TeV)", fontsize=12)

    min_eta = ak.min(
        ak.concatenate(
            [
                seq3_data["nL1BMTFStub"], 
                seq4_data["nL1BMTFStub"], 
                seq5_data["nL1BMTFStub"], 
                data["nL1BMTFStub"]
            ]
        )
    )

    max_eta = ak.max(
        ak.concatenate(
            [
                seq3_data["nL1BMTFStub"], 
                seq4_data["nL1BMTFStub"], 
                seq5_data["nL1BMTFStub"], 
                data["nL1BMTFStub"]
            ]
        )
    )

    bins = np.arange(min_eta, max_eta + 1)

    plt.hist(ak.ravel(data["nL1BMTFStub"]), bins=bins, alpha=0.5, label="full data");
    plt.hist(ak.ravel(seq3_data["nL1BMTFStub"]), bins=bins, alpha=0.5, label="seq3");
    plt.hist(ak.ravel(seq4_data["nL1BMTFStub"]), bins=bins, alpha=0.5, label="seq4");
    plt.hist(ak.ravel(seq5_data["nL1BMTFStub"]), bins=bins, alpha=0.5, label="seq5");

    plt.yscale("log");
    plt.xlabel("nStubs");
    plt.ylabel("Counts");
    plt.legend();
```
